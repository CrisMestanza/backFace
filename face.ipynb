```python
import cv2

camera_id = 1 # o 1, 2, etc. dependiendo de cuál cámara quieras usar

print(f"Conectando a la cámara {camera_id}...")
cap = cv2.VideoCapture(camera_id, cv2.CAP_FFMPEG)  # NOTA: No se usa URL RTSP aquí
cap.set(cv2.CAP_PROP_BUFFERSIZE, 1)

if not cap.isOpened():
    print("No se pudo abrir la cámara.")
else:
    print("Cámara conectada correctamente.")
    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error al leer el frame.")
            break
        cv2.imshow("Cámara Local", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
